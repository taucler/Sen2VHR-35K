# Notebook: Filter Windows by Angle Difference

This notebook filters the window-S2 pairs to keep only those with small viewing zenith angle differences (≤5°). This ensures geometric consistency between VHR and S2 acquisitions for accurate super-resolution training.

Purpose: Apply quality control by removing pairs with significant angle mismatches.

```python
IN_PARQUET = "parquet/windows_best_s2_with_vza.parquet"
THRESH_DEG = 5.0  # degrees
OUT_PARQUET = f"parquet/windows_best_s2_with_vza_angleDiffLE{int(THRESH_DEG)}.parquet"
```

```python
df = pd.read_parquet(IN_PARQUET)
```

```python
# keep windows where |angle_diff_mean| <= 4°
df_f = df[df["angle_diff_mean"].abs() <= THRESH_DEG].copy()
```

```python
print("Original windows:", len(df))
print(f"Kept (|angle_diff_mean| <= {THRESH_DEG}°):", len(df_f))
print("Dropped:", len(df) - len(df_f))
```

```python
df_f.to_parquet(OUT_PARQUET, index=False)
print("Saved:", OUT_PARQUET)
```
